```python
import numpy as np
import random
import math
import h5py
import snapshot as SN
import matplotlib.pyplot as plt
from nbodykit.lab import *
from nbodykit import setup_logging, style
```

```python
basePath = '/Users/usuario-mac/Desktop/Illustris-3'
fields = ['Coordinates']
dm_pos=SN.loadSubset(basePath,135,'DM',fields=fields)
print(dm_pos.shape)
```

```python
position=dm_pos/1000
x=dm_pos[:,0]/1000;
y=dm_pos[:,1]/1000;
z=dm_pos[:,2]/1000;

dm_posArr=np.array([x,y,z],dtype=[('x','f'),('y','f'),('z','f')]).T
print(dm_posArr.shape)
DMdensity_cat = ArrayCatalog(data=dm_posArr)
```

```python
DMdensity_mesh=DMdensity_cat.to_mesh(Nmesh=256,BoxSize=75,window='cic',position='y')
print(DMdensity_cat)

plt.imshow(np.log10(DMdensity_mesh.preview(axes=[0,1])+1))
plt.colorbar()

density=DMdensity_mesh.preview(Nmesh=256)

print(np.mean(density), np.min(density), np.max(density))

f=h5py.File('Density256_DMIll3.hdf5','w')
dset=f.create_dataset('density',data=density)
f.close()
```

```python
filename = 'Density256_DMIll3.hdf5'
f = h5py.File(filename, 'r')
print(f.keys())
data = f['density'][:,:,:]
f.close()

plt.figure()
plt.imshow(np.log10(data[:,100,:]+1))
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(np.log10(data[20,:,:]+1))
plt.colorbar()
```

```python
overdensity=(density-np.mean(density))/np.mean(density)


mesh_over = ArrayMesh(overdensity,BoxSize=75)

r1 = FFTPower(mesh_over, mode='1d', dk=0.5, kmin=0.1)
Pk1 = r1.power

print(Pk1['k'][2], Pk1['power'].real[2])

plt.loglog(Pk1['k'], Pk1['power'].real,'r-')
plt.xlabel(r"k")
plt.ylabel(r"P(k)")
plt.title('P(k) Ill3 DM')
plt.savefig('DM256_power.png')

f1 = open ('power256_DMIll3.dat','w')
save=list(zip(Pk1['k'],Pk1['power'].real))
for h,l in enumerate(save):
    line1=''.join(str(x).ljust(20) for x in l)
    f1.writelines(line1+ '\n')
f1.close()

f5=h5py.File('over256_DMIll3.hdf5','w')
dset=f5.create_dataset('over',data=overdensity)
f5.close()

```
